# Handling Outliers

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
# Load the dataset
df = sns.load_dataset("Iris")
print(df.head())
sns.boxplot(x=df['sepal_width'], data=df)
```

```python
df.shape
```

```python
# IQR
Q1 = np.percentile(df['sepal_width'], 25, method = 'midpoint')
Q3 = np.percentile(df['sepal_width'], 75, method = 'midpoint')

IQR = Q3 - Q1
print("Old Shape: ", df.shape)

Upper = Q3+1.5*IQR
print(Upper)

lower =  Q1-1.5*IQR  
print(lower)

[lower,Upper]
```

```python
print(sns.get_dataset_names())
```

```python
# Upper bound
min_Upper_value = np.where(df['sepal_width'] >= Upper)
print(min_Upper_value) 
```

```python
# Lower bound
max_Lower_value = np.where(df['sepal_width'] <= lower)
print(max_Lower_value)
```

```python
df.drop(max_Lower_value[0], inplace = True)
```

```python
df.drop(min_Upper_value[0], inplace = True)
```

```python
print("New Shape: ", df.shape)
```
